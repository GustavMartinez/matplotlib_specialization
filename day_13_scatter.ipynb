```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

plt.scatter(x,y)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

plt.figure(figsize=(10,6))
plt.scatter(x,y)

plt.xlabel("X values", labelpad= 10)
plt.ylabel("Y values", labelpad= 10)
plt.title("Simple Scatter Plot")

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]
sizes = [100, 200, 300, 400, 500]  # Tamaño de los puntos para darle énfasis

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='tomato', sizes=sizes, alpha=0.7, edgecolor='black')

# Personalización de etiquetas y título
plt.xlabel("X values", fontsize=14, labelpad=10, color='#333333')
plt.ylabel("Y values", fontsize=14, labelpad=10, color='#333333')
plt.title("Enhanced Scatter Plot", fontsize=18, fontweight='bold', pad=15)

# Agregar una cuadrícula para mayor claridad
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Mejorar el estilo de los ejes
plt.xticks(fontsize=12, color='#333333')
plt.yticks(fontsize=12, color='#333333')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
x = np.random.rand(50) * 10   # Genera 50 puntos aleatorios para el eje X
y = np.random.rand(50) * 10   # Genera 50 puntos aleatorios para el eje Y
sizes = np.random.rand(50) * 500  # Genera tamaños aleatorios para los puntos
colors = np.random.rand(50)  # Colores basados en valores aleatorios para el gradiente

# Crear el scatter plot con un gradiente de color
plt.figure(figsize=(12, 8))
scatter = plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='YlGnBu', edgecolor='grey', linewidth=0.5)

# Añadir título y etiquetas con estilo
plt.title("Advanced Scatter Plot with Gradients", fontsize=20, fontweight='bold', color='#333333', pad=20)
plt.xlabel("Random X Values", fontsize=14, labelpad=15, color='#333333')
plt.ylabel("Random Y Values", fontsize=14, labelpad=15, color='#333333')

# Añadir colorbar para el gradiente de colores
cbar = plt.colorbar(scatter)
cbar.set_label('Color Intensity', rotation=270, labelpad=20, fontsize=12)

# Añadir cuadrícula personalizada
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.6, color='grey')

# Mejorar el estilo de los ejes
plt.xticks(fontsize=12, color='#333333')
plt.yticks(fontsize=12, color='#333333')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Añadir anotaciones para puntos específicos
for i in range(len(x)):
    if i % 10 == 0:  # Añadir anotación en cada 10º punto para no sobrecargar el gráfico
        plt.text(x[i], y[i], f'({x[i]:.1f}, {y[i]:.1f})', fontsize=10, ha='right', color='black')

# Ajuste de márgenes para un mejor equilibrio visual
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
