```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
line1, = plt.plot([1,2,3,4,5])
line2, = plt.plot([3,4,5,6,7])

plt.legend([line1, line2], ['Line 1', 'Line 2'], loc=(1.05, 0), markerfirst=False, shadow=True, frameon=True, ncol=2, columnspacing = 5)

# markerfirst = boolean to change the order of the legend.
# shadow = shows legend's shadow.
# frameon = boolean that shows the border of the legend
# columnspacing = change the space between the elements of the column


plt.show()
```

### Title to the legend

```python
line1, = plt.plot([1,2,3,4,5])
line2, = plt.plot([3,4,5,6,7])

plt.legend([line1, line2], ['Line 1', 'Line 2'], loc=(1.05, 0), title= "Legend", title_fontsize=20, borderpad = 5, labelspacing = 5)


# add a title to the legend
# title_fontsize change the size of the title
# borderpad = border padding
# labelspacing = space between the legend entries


plt.show()
```

### Ticks location

```python
line1, = plt.plot([1,2,3,4,5])
line2, = plt.plot([3,4,5,6,7])
line3, = plt.plot([7,8,9,10,11])
line4, = plt.plot([6,5,4,3,2])

plt.legend([line1, line2], ['Line 1', 'Line 2'])

plt.grid()
plt.xticks([0,1,2,3,4], ["A", "B", "C", "D", "E"], color="red", size=20, rotation=45)

plt.show()
```
