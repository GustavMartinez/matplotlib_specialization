```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [13, 82, 50, 6, 40, 92, 73, 84, 82, 5, 56, 5, 67, 100, 20, 3, 45, 91, 38, 41, 
          65, 48, 34, 64, 51, 12, 68, 52, 69, 65, 99, 55, 95, 53, 8, 22, 10, 53, 72, 
          35, 76, 73, 20, 70, 64, 57, 74, 30, 97, 74, 140]

plt.boxplot(data)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data, 
            patch_artist=True,  # Llena el cuadro con color
            boxprops=dict(facecolor='lightblue', color='darkblue', linewidth=1.5),  # Propiedades de la caja
            medianprops=dict(color='darkred', linewidth=2),  # Estilo de la mediana
            whiskerprops=dict(color='darkblue', linewidth=1.5),  # Estilo de las líneas
            capprops=dict(color='darkblue', linewidth=1.5),  # Estilo de las líneas horizontales de los límites
            flierprops=dict(marker='o', color='red', markersize=6))  # Estilo de los valores atípicos

plt.title("Box Plot Intermedio", fontsize=16, fontweight='bold')
plt.ylabel("Valores", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

box = plt.boxplot(data, 
                  patch_artist=True,  # Llena el cuadro con color
                  boxprops=dict(facecolor='skyblue', color='darkblue', linewidth=1.5, alpha=0.6),  # Cuadro
                  medianprops=dict(color='orange', linewidth=2),  # Mediana
                  whiskerprops=dict(color='darkblue', linewidth=1.5, linestyle='--'),  # Bigotes
                  capprops=dict(color='darkblue', linewidth=1.5),  # Límites
                  flierprops=dict(marker='D', color='red', markersize=8, alpha=0.7),  # Valores atípicos
                  widths=0.5)  # Ancho del cuadro

plt.title("Box Plot Avanzado: Distribución de Datos", fontsize=18, fontweight='bold', color='darkblue')
plt.ylabel("Valores", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Añadir etiquetas para la mediana, valores mínimos y máximos
median_val = np.median(data)
min_val = np.min(data)
max_val = np.max(data)

plt.text(1.1, median_val, f'Mediana:\n{median_val}', va='center', ha='left', color='orange', fontsize=12)
plt.text(1.1, min_val, f'Mínimo:\n{min_val}', va='center', ha='left', color='darkblue', fontsize=12)
plt.text(1.1, max_val, f'Máximo:\n{max_val}', va='center', ha='left', color='darkblue', fontsize=12)

# Identificar valores atípicos y anotarlos
outliers = [point.get_ydata()[0] for point in box['fliers']]  # Obtiene los valores de los outliers
for outlier in outliers:
    plt.annotate(f'{outlier}', (1.05, outlier), textcoords="offset points", xytext=(5,-5), ha='center', fontsize=11, color='red')

plt.show()
```
