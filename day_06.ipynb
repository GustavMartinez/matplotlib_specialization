```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4,5,4,3,2,1], 'b+', label='blue line')
plt.plot([8,5,4,6,7,4,6,9,6], 'k', label='black line')

plt.grid(alpha=0.3)
plt.xlim(-1,9)
plt.ylim(0,9)

plt.legend()


plt.show()
```

```python
blue_line, = plt.plot([1,2,3,4,5,4,3,2,1], 'b+')
black_line, = plt.plot([8,5,4,6,7,4,6,9,6], 'k')

plt.grid(alpha=0.3)

plt.legend([blue_line, black_line], ['Blue Line', 'Black Line'])


plt.show()
```

```python
line1, = plt.plot([1,2,3,4,5], 'b')
line2, = plt.plot([5,4,3,2,1], 'c')

plt.grid(alpha=0.2)

plt.legend([line1, line2], ['Linea Azul', 'Linea Cian'], loc=7) #loc >> localización de la leyenda

"""
Legend position: (loc=)

best - 0
upper right - 1
upper left - 2
lower left - 3
lower right - 4
right - 5
center left - 6
center right - 7
lower center - 8
upper center - 9
center - 10
"""

plt.show()
```

```python
line1, = plt.plot([1,2,3,4,5], 'b')
line2, = plt.plot([5,4,3,2,1], 'c')

plt.grid(alpha=0.2)

plt.legend([line1, line2], ['Linea Azul', 'Linea Cian'], loc=8, ncol=2) #ncol >> numero de columnas


plt.show()
```
