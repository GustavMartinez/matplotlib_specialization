```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4,5,4,3,2,1], 'b+', label='blue line')
plt.plot([8,5,4,6,7,4,6,9,6], 'k', label='black line')

plt.grid(alpha=0.3)
plt.xlim(-1,9)
plt.ylim(0,9)

plt.legend()


plt.show()
```

```python
blue_line, = plt.plot([1,2,3,4,5,4,3,2,1], 'b+')
black_line, = plt.plot([8,5,4,6,7,4,6,9,6], 'k')

plt.grid(alpha=0.3)

plt.legend([blue_line, black_line], ['Blue Line', 'Black Line'],loc=8)


plt.show()
```

```python
line1, = plt.plot([1,2,3,4,5], 'b')
line2, = plt.plot([5,4,3,2,1], 'c')

plt.grid(alpha=0.2)

plt.legend([line1, line2], ['Linea Azul', 'Linea Cian'], loc=7) #loc >> localización de la leyenda

"""
Legend position: (loc=)

best - 0
upper right - 1
upper left - 2
lower left - 3
lower right - 4
right - 5
center left - 6
center right - 7
lower center - 8
upper center - 9
center - 10
"""

plt.show()
```

### Legend in columns

```python
line1, = plt.plot([1,2,3,4,5], 'b')
line2, = plt.plot([5,4,3,2,1], 'c')

plt.grid(alpha=0.2)

plt.legend([line1, line2], ['Linea Azul', 'Linea Cian'], loc=8, ncol=2) #ncol >> numero de columnas


plt.show()
```

### Adjust the legend location (tuple)

```python
plt.plot([1,2,3,4,5], label="Azul")
plt.plot([3,4,5,6,7], label="Naranja")
plt.plot([5,6,7,8,9], label="Verde")

plt.grid(alpha=0.2)


# Posicionar la leyenda para que quede en el punto x=3 y y=1. El punto de referencia es la esquina inferior izquierda
plt.xlim(0,4)
plt.ylim(0,9)
plt.legend(loc=(0.5/4,7/9))
# los tres comandos anteriores son importantes para el posicionamiento


plt.show()
```

### Bbox - 4 values

```python
plt.plot([1,2,3,4,5], label="Azul")
plt.plot([3,4,5,6,7], label="Naranja")
plt.plot([5,6,7,8,9], label="Verde")

plt.grid(alpha=0.2)

##bbox crea un espacio donde será posicionada la leyenda y, dentro de este, puede cambiarse la posición (loc)

plt.xlim(0,4)
plt.ylim(0,9)

plt.legend(bbox_to_anchor=(0.5,0/9,0.5,0.5), loc='lower right')

# bbox_to_anchor = (x, y, width, hight)


plt.show()
```

### Legend outside the plot

```python
plt.plot([1,2,3,4,5], label="Azul")
plt.plot([3,4,5,6,7], label="Naranja")
plt.plot([5,6,7,8,9], label="Verde")

plt.grid(alpha=0.2)

##bbox crea un espacio donde será posicionada la leyenda y, dentro de este, puede cambiarse la posición (loc)

plt.xlim(0,4)
plt.ylim(0,9)

plt.legend(loc=(1.05,0))



plt.show()
```

### Font size and color of the legend

```python
plt.plot([1,2,3,4,5], label="Azul")
plt.plot([3,4,5,6,7], label="Naranja")
plt.plot([5,6,7,8,9], label="Verde")

plt.grid(alpha=0.2)

##bbox crea un espacio donde será posicionada la leyenda y, dentro de este, puede cambiarse la posición (loc)

plt.xlim(0,4)
plt.ylim(0,10)

plt.legend(loc=(1.05,0.5), fontsize=20, facecolor='red', edgecolor='black', framealpha=0.3)

# fontsize = tamaño de la fuente
# facecolor = Background of the legend
# edgecolor = color of the border
# framealpha = Transparency of the legend


plt.show()
```
