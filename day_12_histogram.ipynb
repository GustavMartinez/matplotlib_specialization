```python
# Aplicar un estilo de fondo predefinido (opcional).
plt.style.use('ggplot')

# Valores de edad para el histograma
edades = [13, 82, 50, 6, 40, 92, 73, 84, 82, 5, 56, 5, 67, 100, 20, 3, 45, 91, 38, 41, 
          65, 48, 34, 64, 51, 12, 68, 52, 69, 65, 99, 55, 95, 53, 8, 22, 10, 53, 72, 
          35, 76, 73, 20, 70, 64, 57, 74, 30, 97, 74]

# Crear el histograma
plt.figure(figsize=(10, 6))

plt.hist(edades, bins=10, color='#129595', edgecolor='black', alpha=0.6)

plt.title("Distribución de Edades", fontsize=18, fontweight='bold')
plt.xlabel("Edad", fontsize=14)
plt.ylabel("Frecuencia", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Agregar cuadrícula horizontal para facilitar lectura
# - 'axis="y"' limita la cuadrícula al eje Y

plt.show()
```

```python
plt.style.use('ggplot')

edades = [13, 82, 50, 6, 40, 92, 73, 84, 82, 5, 56, 5, 67, 100, 20, 3, 45, 91, 38, 41, 
          65, 48, 34, 64, 51, 12, 68, 52, 69, 65, 99, 55, 95, 53, 8, 22, 10, 53, 72, 
          35, 76, 73, 20, 70, 64, 57, 74, 30, 97, 74]

plt.figure(figsize=(10, 6))

n, bins, patches = plt.hist(edades, bins=10, color='#129595', edgecolor='black', alpha=0.6)


# Agregar anotaciones en cada barra del histograma
for i in range(len(patches)):
    plt.text(patches[i].get_x() + patches[i].get_width() / 2, patches[i].get_height(),
             int(patches[i].get_height()), ha='center', va='bottom', color='black', fontweight='bold')
    # - 'plt.text' coloca el texto encima de cada barra
    # - 'get_x() + get_width() / 2' centra el texto en la barra
    # - 'get_height()' obtiene la altura (frecuencia) para posicionar el valor
    # - 'fontweight' en 'bold' resalta el valor de la frecuencia


plt.title("Distribución de Edades", fontsize=18, fontweight='bold')
plt.xlabel("Edad", fontsize=14)
plt.ylabel("Frecuencia", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7) 

plt.show()
```
