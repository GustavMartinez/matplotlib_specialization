```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line graph

```python
year = [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010]
unemply = [9.8,12,8,7,5.6,6.8,7,8,6,4]

plt.plot(year, unemply, 'bo-.')

plt.grid(alpha=0.2)

plt.xlabel("Year")
plt.ylabel("Unemployment rate")
plt.title("Year vs unemployment rate")


plt.show()
```

### Bar chart - Bar graph

```python
x = ["science", 'commerce', 'arts']
h = [200,300,500]
w = [0.5,0.7,0.9]

c = ['red', 'green', 'chocolate']

plt.bar(x, h, width=w, bottom=100, color=c)

plt.grid(alpha=0.2)

"""
change the width of the bars >> width
change the start of the bars >> bottom
change the colors of the bars >> color
"""

plt.show()
```

### Horizontal bar chart

```python
y = ["science", 'commerce', 'arts']
w = [200,300,500]
h = [0.5,0.7,0.9]

c = ['red', 'green', 'chocolate']

plt.barh(y, w, 0.5, color=c)

plt.grid(alpha=0.2)

"""
change the width of the bars >> width
change the start of the bars >> bottom
change the colors of the bars >> color
"""

plt.show()
```

### Stacked Bar Graph

```python
countries = ['USA', "Russia", "GBR", "China", "Germany"]

gold = [20,10,6,24,30]
silver = [100,50,30,120,150]
bronze = [200,100,60,240,300]

bottom_bronze = np.add(gold, silver)

plt.bar(countries, gold, 0.4, label = "Gold")
plt.bar(countries, silver, 0.4,bottom=gold, label = "Silver")
plt.bar(countries, bronze, 0.4, bottom=bottom_bronze, label = "Bronze")

plt.legend()

plt.show()
```
