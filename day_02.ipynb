```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2,3,5,7,11], 'rD', markersize=15) #format: color, marker, marker size
plt.plot([7,4,3,2,1], 'b:', linewidth=5) # format: color, line type, line width

plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("First graph")

plt.show()
```

```python
# Markers:
"""
. point
, pixel
o circle
v, ^, <, > triangle
1, 2, 3, 4 tri
s square
p pentagon
* star
h, H hexagon
+ plus
x x_marker
D, d diamond
| vline
_ hline

"""

# Line styles

"""
- solid
-- dashed
-. dash-dot
: dotted

"""
```

```python
a = np.arange(0,10,0.5)

plt.plot(a, "go:", a**3, "r^--", a**2, 'c+-.')


plt.title("Second Plot")
plt.xlabel("X label")
plt.ylabel("Y label")

plt.show()
```

## Keyword arguments

```python
plt.plot([10,20,30,40], '--', linewidth=10, color='red')

plt.show()
```

```python
plt.plot([10,20,30,40], color= 'green', dashes=[7,3]) #dashes: [line, space]

plt.show()
```

```python
plt.plot([10,20,30,40], color= 'green', linestyle=":")
plt.plot([35,25,15,10], color= 'red', linestyle="-.")
plt.plot([15,30,40,20], color='y', linestyle="dashdot")

plt.show()
```
