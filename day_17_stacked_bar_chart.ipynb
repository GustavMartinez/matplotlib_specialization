```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
countries = ['USA', "Russia", "GBR", "China", "Germany"]

gold = [20,10,6,24,30]
silver = [100,50,30,120,150]
bronze = [200,100,60,240,300]

bottom_bronze = np.add(gold, silver)

plt.bar(countries, gold, 0.4, label = "Gold")
plt.bar(countries, silver, 0.4,bottom=gold, label = "Silver")
plt.bar(countries, bronze, 0.4, bottom=bottom_bronze, label = "Bronze")

plt.legend()

plt.show()
```

```python
countries = ['USA', "Russia", "GBR", "China", "Germany"]
gold = [20, 10, 6, 24, 30]
silver = [100, 50, 30, 120, 150]
bronze = [200, 100, 60, 240, 300]

# Clculo para apilar barras
bottom_bronze = np.add(gold, silver)

# Configuracin del grfico
plt.figure(figsize=(10, 6))

# Barras apiladas
plt.bar(countries, gold, 0.4, color='gold', edgecolor='black', linewidth=0.5, label="Gold")
plt.bar(countries, silver, 0.4, bottom=gold, color='silver', edgecolor='black', linewidth=0.5, label="Silver")
plt.bar(countries, bronze, 0.4, bottom=bottom_bronze, color='#cd7f32', edgecolor='black', linewidth=0.5, label="Bronze")

# Ttulo y etiquetas
plt.title("Medallas por Pas", fontsize=16, fontweight='bold', pad=20, color='darkblue')
plt.xlabel("Pases", fontsize=12, labelpad=10, color='darkblue')
plt.ylabel("Cantidad de Medallas", fontsize=12, labelpad=10, color='darkblue')

# Personalizacin de ticks
plt.xticks(fontsize=10, color='darkblue')
plt.yticks(fontsize=10, color='darkblue')

# Lneas de cuadrcula para mejor visualizacin
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Leyenda con estilo
plt.legend(title="Tipos de Medalla", fontsize=12, loc=2, frameon=True, facecolor='lightgray', edgecolor='black')

# Mostrar el grfico
plt.tight_layout()
plt.show()
```
