```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [1,4,3,4,5,4,3,2]
x = [1,2,3,4,5,6,7,8]

plt.fill_between(x,y) # área del gráfico

plt.show()
```

```python
y = [1,4,3,4,5,4,3,2]
x = [1,2,3,4,5,6,7,8]

plt.fill_between(x,y, color='lightblue') #Área del gráfico
plt.plot(x,y, color='blue', alpha=0.5) # Linea del gráfico

plt.xlabel("X label", labelpad=10)
plt.ylabel("y label", labelpad=10)
plt.title("Area Graph", pad=15)

plt.xlim(1,8)
plt.ylim(0,6)

plt.grid(alpha=0.2)

plt.show()
```

```python
y = [1, 4, 3, 4, 5, 4, 3, 2]
x = [1, 2, 3, 4, 5, 6, 7, 8]

plt.figure(figsize=(10, 6))

plt.fill_between(x, y, color='skyblue', alpha=0.7, label='Área Rellena')
plt.plot(x, y, color='darkblue', alpha=0.6, linewidth=1.2, linestyle='--', label='Línea Principal')

plt.xlim(1, 8)
plt.ylim(0, 6)
plt.grid(alpha=0.3, linestyle='--', linewidth=0.8)

plt.xlabel("Eje X", labelpad=10, fontsize=12, color='darkblue', fontstyle='italic')
plt.ylabel("Eje Y", labelpad=10, fontsize=12, color='darkblue', fontstyle="italic")
plt.title("Gráfico de Áreas con Estilo Intermediario", pad=15, fontsize=16, fontweight='bold', color='navy', loc='right')


plt.xticks(fontsize=10, color='darkblue')
plt.yticks(fontsize=10, color='darkblue')

plt.legend(fontsize=12, loc=2, frameon=False)

plt.show()
```

```python
y = [1, 4, 3, 4, 5, 4, 3, 2]
x = [1, 2, 3, 4, 5, 6, 7, 8]

plt.figure(figsize=(12, 7))

plt.fill_between(x, y, color='skyblue', alpha=0.5, label='Área Rellena')
plt.plot(x, y, color='darkblue', alpha=0.9, linewidth=1.5, linestyle='--', label='Tendencia')

plt.xlim(0.5, 8.5)
plt.ylim(0, 6)
plt.grid(alpha=0.4, linestyle='-.', linewidth=0.7, color='gray')

plt.xlabel("Eje X (Categorías)", labelpad=15, fontsize=14, color='darkblue', fontstyle='italic')
plt.ylabel("Valores del Eje Y", labelpad=15, fontsize=14, color='darkblue', fontstyle="italic")
plt.title("Gráfico de Áreas Avanzado", pad=20, fontsize=18, fontweight='bold', color='darkred', loc='center')

plt.xticks(ticks=x, labels=[f'P{x}' for x in x], fontsize=12, color='darkblue', rotation=20)
plt.yticks(fontsize=12, color='darkblue')

plt.legend(fontsize=13, loc='upper left', frameon=True, facecolor='lightgray', edgecolor='gray', shadow=True)

for xi, yi in zip(x, y):
    plt.text(xi, yi + 0.2, f'{yi}', ha='center', va='bottom', fontsize=10, color='darkblue')

avg_y = sum(y) / len(y)
plt.axhline(avg_y, color='red', linestyle='--', linewidth=1.5, label='Promedio')
plt.text(7.7, avg_y, f'Promedio: {avg_y:.1f}', fontsize=11, color='red', va='bottom')

plt.tight_layout()
plt.show()
```
