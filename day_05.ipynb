```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,4,8,16,32,64,128], 'c+-.', markeredgecolor='b', markeredgewidth=2)


plt.xlabel("X label", color='blue', weight=800, size=15, labelpad=15, style='italic')
plt.ylabel("Y label", color='blue', weight=800, size=15, labelpad=15, style='italic')

plt.title("Graph title", color='black', weight=800, style='italic', size=15, loc='right', pad=25)

"""
position of the title : loc

'right'
'left'
'center'
"""

"""
space between text title and graph: pad

"""


plt.show()
```

### Grid

```python
plt.plot([10,20,30], [1,2,3])

plt.grid()

plt.show()
```

```python
plt.plot([1,2,3,4])

plt.minorticks_on()
plt.grid(color='red', which='minor')

plt.show()
```

```python
plt.plot([1,2,3,4])

plt.grid(color='red', which='major', axis='x', alpha=0.3)

plt.show()
```

### x and y axis limits

```python
plt.plot([1,2,3,4,5], 'k')


"""
xlim/ylim define the limit of the right and left side of the x axis. this function takes two values

"""

plt.xlim(0,4)
plt.ylim(0,6)


plt.grid(alpha=0.3)

plt.show()
```
