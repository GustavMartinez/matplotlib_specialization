```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]  # Porcentajes para cada sección
colors = ['gold', 'lightblue', 'lightcoral', 'lightgreen']  # Colores para cada sección

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# - autopct='%1.1f%%' muestra el porcentaje de cada sección con una precisión de un decimal.
# - startangle=90 rota el gráfico para empezar desde el ángulo de 90 grados 

plt.show()
```

```python
# Datos de ejemplo
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]  # Porcentajes para cada sección
colors = ['#ffcc00', '#66b3ff', '#ff6666', '#99ff99']  # Colores personalizados

# Explosión de una sección (hacer que una sección se "explode" o sobresalga)
explode = (0.1, 0, 0, 0)  # Explota la primera sección (A)

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, 
        explode=explode, shadow=True, wedgeprops={'edgecolor': 'black', 'linewidth': 1})

# Añadir título con estilo
plt.title("Gráfico de Pastel Intermedio", fontsize=18, fontweight='bold', color='darkblue')

# Mejorar la visualización
plt.axis('equal')  # Asegura que el pastel sea un círculo perfecto

# Mostrar el gráfico
plt.show()
```
