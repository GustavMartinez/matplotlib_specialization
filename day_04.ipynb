```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([5,10,15,23,24,26], 'r*--')

#change label properties with a dictionary
plt.xlabel("x label", {"size":20, "color":"blue"}, labelpad=25) #labelpad: distance from the line of the label to the label name

plt.show()
```

```python
plt.plot([5,10,15,23,24,26], 'r*--')

#change label properties with a dictionary
plt.xlabel("x label", labelpad=25, size=30, color='c') #labelpad: distance from the line of the label to the label name

plt.show()
```

```python
plt.plot([5,10,15,23,24,26], 'rv--')

#change label properties with a dictionary
plt.ylabel("y label", labelpad=25, size=30, color='c') #labelpad: distance from the line of the label to the label name

plt.show()
```

```python
plt.plot([5,10,15,23,24,26], 'rv--')

#change label properties with a dictionary
plt.ylabel("y label", labelpad=25, size=30, color='c') #labelpad: distance from the line of the label to the label name
plt.xlabel("x label", labelpad=10, size=15, color='r', style='italic')



"""
FONTSTYLE:

'normal'
'italic'
'oblique'

"""


plt.show()
```

```python
plt.plot([5,10,15,23,24,26], 'rv--')

#change label properties with a dictionary
plt.ylabel("y label", labelpad=25, size=30, color='c', weight=800) #labelpad: distance from the line of the label to the label name
plt.xlabel("x label", labelpad=10, size=15, color='r', style='italic')



""" 
fontweight or weight

a numerical value in range 0 - 1000

"""
plt.show()
```

```python
plt.plot([5,10,15,23,24,26], 'rv--')

#change label properties with a dictionary
plt.ylabel("y label", labelpad=25, size=30, color='c', weight=800, fontfamily='serif') #labelpad: distance from the line of the label to the label name
plt.xlabel("x label", labelpad=10, size=15, color='r', style='italic', fontfamily='serif')

"""
fontfamily:

serif
sans-serif
monospace
..etc

"""

plt.show()
```
