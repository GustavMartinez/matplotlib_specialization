```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.plot([8,7,6,5,6,7,8], marker= "+", markeredgecolor="red") # The first defauld values are blue color and solid line.
plt.plot([5,6,7,8,7,6,5], marker= "+", markeredgecolor="red", markeredgewidth=20) # The second defauld values are blue color and solid line.
""" Markeredgewidth changes the thikness of the marker"""
plt.plot([6,7,8,9,8,7,6], marker='h', markersize=20, markerfacecolor="k") # Markerfacecolor changes the front color of the marker
""" Markersize changes the size of the marker"""

plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("First graph")

plt.show()
```

```python
plt.plot([1,2,4,8,16,32,64], marker='d', markevery=2)
""" Skips a step"""
plt.show()
```

```python
plt.plot([1,2,3], [4,5,6], alpha=0.5, linewidth=15) #alpha to adjust the transparency. range between 1 and 0
plt.show()
```

```python
""" Order of the plot"""

plt.plot([10,20,80,90], linewidth=30,zorder=2) # Zorder indicates which data will be displayed at front
plt.plot([10,20,30,40], linewidth=30,zorder=1)

plt.show()
```
